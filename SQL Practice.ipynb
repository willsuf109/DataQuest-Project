```python
%%sql
WITH named_table AS
    (SELECT  Team_Name, (38) Pl, W,D,L,F,A, (F - A) GD, (W*3 + D) AS pts
    FROM Tbl_results tr
    LEFT JOIN Tbl_team tt ON tr.Team_Id = tt.Team_Id
    ORDER BY pts DESC)
    
SELECT ROW_NUMBER() OVER() AS Place,*
    FROM named_table

```
